Data Visualization 

```python
plt.figure(figsize=(15,10))
country=data_new['country_txt'].value_counts().index
sns.countplot(data_new['country_txt'],order=country[:10])
plt.xlabel("Countries")
plt.ylabel("Count of terrorist attacks")
plt.title("Top 10 countries having highest number of terrorist attacks")
```

```python
plt.figure(figsize=(25,10))
region=data_new['region_txt'].value_counts().index
sns.countplot(data_new['region_txt'],order=region[:10])
plt.xlabel("Regions")
plt.ylabel("Count of terrorist attacks")
plt.title("Top 10 regions having highest number of terrorist attacks")
```

```python
plt.figure(figsize=(15,10))
city=data_new['city'].value_counts().index
sns.countplot(data_new['city'],order=city[:10])
plt.xlabel("cities")
plt.ylabel("Count of terrorist attacks")
plt.title("Top cities having highest number of terrorist attacks")
```

```python
plt.figure(figsize=(15,10))
year=data_new['iyear'].value_counts().index
sns.countplot(data_new['iyear'],order=year[:10])
plt.xlabel("Year")
plt.ylabel("Count of terrorist attacks")
plt.title("The year in which the top 10 countries having highest number of terrorist attacks")
```

```python
plt.figure(figsize=(25,10))
target=data_new['targtype1_txt'].value_counts().index
sns.countplot(data_new['targtype1_txt'],order=target[:10])
plt.xlabel("Type of property targeted")
plt.ylabel("Count of terrorist attacks")
plt.title("Top 10 categories targeted by terrorists")
```

```python
plt.figure(figsize=(50,15))
wepon=data_new['weaptype1_txt'].value_counts().index
sns.countplot(data_new['weaptype1_txt'],order=wepon[:10])
plt.xlabel("Weapons")
plt.ylabel("Number of times the weapons been used by terrorists.")
plt.title("Top 5 categories of weapons used by the terrorists.")
```

```python
plt.figure(figsize=(20,10))
group_name=data_new['gname'].value_counts().index
sns.countplot(data_new['gname'],order=group_name[:5])
plt.xlabel("Gangsters")
plt.ylabel("Count of gangsters attacks")
plt.title("Top 5 categories of terrorist groups")
```

```python
data_order=data_new.groupby('country_txt')['extended'].sum()
data_order.head()
```

```python
data_order=pd.DataFrame(data_order)
data_order=data_order.sort_values('extended',ascending=False)
data_order.head()
```

```python
data_order=data_order.reset_index()
data_order.head()
```
